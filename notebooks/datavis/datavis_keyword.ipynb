# Import libraries and variables

```python
import json
import pandas as pd
import os
from sortedcontainers import SortedDict

%store -r output_activity_dir
%store -r activity_all_file
%store -r output_events_nonmoma

common_english_words = ["a","able","about","across","after","all","almost","also","am","among","an","and","any","are","as","at","be","because","been","but","by","can","cannot","could","dear","did","do","does","either","else","ever","every","for","from","get","got","had","has","have","he","her","hers","him","his","how","however","i","if","in","into","is","it","its","just","least","let","like","likely","may","me","might","most","must","my","neither","no","nor","not","of","off","often","on","only","or","other","our","own","rather","said","say","says","she","should","since","so","some","than","that","the","their","them","then","there","these","they","this","tis","to","too","twas","us","wants","was","we","were","what","when","where","which","while","who","whom","why","will","with","would","yet","you","your"]

list_remove = [',', '!', '.', ':', '(', ')']
```

# Keyword frequency in exhibition title - all organisations - Top 50


```python

numbers = {}

with open(os.path.join(output_activity_dir, activity_all_file), 'r') as json_file:
    
    data = json.load(json_file)
    # iterate over events
    for event in data.get("events"):
        
        # event label
        title = event.get("_label")
       
        keywords = title.split(" ")
        for keyword in keywords:
            keyword = ''.join(char for char in keyword if char not in list_remove)
            if keyword.lower() in common_english_words:
                continue
            if keyword in numbers:
                numbers[keyword] += 1
            else:
                numbers[keyword] = 1
    
    numbers = dict(sorted(numbers.items(), key=lambda item: item[1]))
    
    df = pd.DataFrame(numbers.items()).tail(50)
    df.columns=["Word", "Count"]
    display(df)
```

# Keyword frequency in exhibition title - non-MoMA - Top 50

```python


numbers = {}

with open(os.path.join(output_activity_dir, output_events_nonmoma), 'r') as json_file:
 
    data = json.load(json_file)
    for event in data.get("events"):   
        title = event.get("_label")
       
        keywords = title.split(" ")
        for keyword in keywords:
            keyword = ''.join(char for char in keyword if char not in list_remove)
            if keyword.lower() in common_english_words:
                continue
            if keyword  in numbers:
                numbers[keyword] += 1
            else:
                numbers[keyword] = 1
    
    numbers = dict(sorted(numbers.items(), key=lambda item: item[1]))
    
        
    df = pd.DataFrame(numbers.items()).tail(50)
    df.columns=["Word", "Count"]
    display(df)
        
```
