# Activity - data analysis

```python
try:
    import json
except:
    %pip install json
    import json

try:
    import requests
except:
    %pip install requests
    import requests

try:
    import pandas as pd
except:
    %pip install pandas
    import pandas as pd

from operator import itemgetter

import os
```

```python


filepath = "data/activity/"

exhibitions = []

exhibition_template = {
    "id": "",
    "label": "",
    "start": "",
    "end": "",
    "location": "",
    "org": "",
    "venue": "", 
    "_label":"",
    "identified_by": [],
    "timespan": [],
    "carried_out_by": [],
    "influenced_by": []
}


for filename in os.listdir(filepath):
    try:
        with open(os.path.join(filepath, filename), 'r') as json_file:
            
            data = json.load(json_file)

            this_exhibition = exhibition_template.copy()
            this_exhibition["id"] = data.get("id")
            this_exhibition["label"] = data.get("_label")
            this_exhibition["_label"] = data.get("_label")

            this_exhibition["start"] = data.get("timespan").get("begin_of_the_begin")
            this_exhibition["end"] = data.get("timespan").get("end_of_the_end")

            this_exhibition["location"] = data.get("took_place_at")[0].get("defined_by")
            this_exhibition["venue"] = data.get("took_place_at")[0].get("_label")
            
            this_exhibition["org"] = data.get("carried_out_by")[0].get("_label")
            
            
            carried_out = []
            
            for ag in data.get("carried_out_by"):
                carried_out.append(ag.get("id"))
            this_exhibition["carried_out_by"] = carried_out
            

            influencers = []
            for inf in data["influenced_by"]:
                influencers.append(inf.get("id"))

            this_exhibition["influenced_by"] = influencers

            exhibition_copy = this_exhibition.copy()
            exhibitions.append(exhibition_copy) 
            
    except:
        continue



with open('api/events_all.json', 'w') as ex_file:
     ex_file.write(json.dumps({"events": exhibitions}))
```

```python

with open('api/events_all.json', 'r') as exjson_file:
    exdata = json.load(exjson_file)

    ex = exdata.get("events")

    # order by event._label asc
    sorted_list_title_asc = sorted(exdata.get('events'), key=itemgetter('_label'), reverse=False) 
    sliced_list_title_asc = sorted_list_title_asc[:200]
    # write to file 
    with open('api/events_title_asc.json', 'w') as file:
        file.write(json.dumps({"events": sliced_list_title_asc}))

    # order by event._label dsc
    sorted_list_title_dsc = sorted(exdata.get('events'), key=itemgetter('_label'), reverse=True) 
    sliced_list_title_dsc = sorted_list_title_dsc[:200]
    # write to file 
    with open('api/events_title_dsc.json', 'w') as file:
        file.write(json.dumps({"events": sliced_list_title_dsc}))

    # .start asc 
    sorted_list_start_asc = sorted(exdata.get('events'), key=itemgetter('start'), reverse=False) 
    sliced_list_start_asc = sorted_list_start_asc[:200]
    # write to file 
    with open('api/events_start_asc.json', 'w') as file:
        file.write(json.dumps({"events": sliced_list_start_asc}))
        
    # order by start dsc
    sorted_list_start_dsc = sorted(exdata.get('events'), key=itemgetter('start'), reverse=True) 
    sliced_list_start_dsc = sorted_list_start_dsc[:200]
    # write to file 
    with open('api/events_start_dsc.json', 'w') as file:
        file.write(json.dumps({"events": sliced_list_start_dsc}))

   
 
    
   
       
       
```
